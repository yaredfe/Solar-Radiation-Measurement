```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv("C:\\Users\\User\\Downloads\\film\\Solar-Radiation-Measurement\\notebooks\\new_data_sets\\cleaned_data_for_benin-malanville.csv")
```

```python
data.drop(columns="Unnamed: 0",axis=1)
```

```python
plt.figure(figsize=(14, 9))

# Plot each feature
plt.plot(data['GHI'].iloc[:1000], label='GHI')  # Plot GHI data
plt.plot(data['DNI'].iloc[:1000], label='DNI')  # Plot DNI data
plt.plot(data['DHI'].iloc[:1000], label='DHI')  # Plot DHI data
plt.plot(data['Tamb'].iloc[:1000], label='Tamb')  # Plot Tamb data

# Add titles and labels
plt.title('Time Series of Solar Irradiance and Temperature')
plt.xlabel('Time')
plt.ylabel('Values')

# Add legend
plt.legend()

# Set x-ticks at every 10th data point
tick_positions = np.arange(0, len(data['GHI'].iloc[:1000]),30)  # Adjust the interval as needed
plt.xticks(tick_positions, rotation=50)  # Rotate x-axis labels by 50 degrees for better readability

# Display the plot
plt.show()

# Save the plot to a file
plt.savefig("time_series5.png")
```

```python
sampled_df = data.sample(frac=0.1, random_state=1)
```

```python
corr = sampled_df[['GHI', 'DNI', 'DHI', 'TModA', 'TModB']].corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
plt.title('Correlation Heatmap of Solar Radiation and Temperature Measures')
plt.show()
plt.savefig("5.1heatmap.jpg")
```

```python
sns.pairplot(sampled_df[['GHI', 'DNI', 'DHI', 'TModA', 'TModB']])
plt.suptitle('Pair Plot of Solar Radiation and Temperature Measures', y=1.02)
plt.show()
plt.savefig("5.1pair plot.jpg")
```

```python
sns.pairplot(sampled_df[['GHI', 'DNI', 'DHI', 'WS', 'WSgust', 'WD']])
plt.suptitle('Scatter Matrix of Wind Conditions and Solar Irradiance', y=1.02)
plt.show()
plt.savefig("5.1scatter matric.jpg")
```
